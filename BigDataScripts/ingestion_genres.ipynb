```python
import findspark
# Locate the Spark installation
findspark.init()

import pyspark
from pyspark.sql.functions import col
from pyspark.sql import *

spark = SparkSession.builder\
        .appName("IngestionG")\
        .master("local[*]")\
        .config("spark.driver.memory", "5g")\
        .config("spark.executor.memory", "4g")\
        .config("spark.storage.memoryFraction", "0.5")\
        .config("spark.shuffle.memoryFraction", "0.5")\
        .config("spark.driver.maxResultSize", "0")\
        .getOrCreate()
```

```python
from pyspark.sql.types import StructType, StructField, StringType, LongType

# Define the schema
schema = StructType([
    StructField('movieId', LongType(), True),
    StructField('title', StringType(), True),
    StructField('genres', StringType(), True)])

# Read the CSV file from a directory
df = spark.read.csv('hdfs://localhost:54310/user/reviews/kaggledata/movie.csv', schema=schema).select("title", "genres")
df.limit(5).toPandas()
```

```python
# Save the DataFrame as CSV in HDFS
df.write.csv(path='hdfs://localhost:54310/user/reviews/data/genre', mode='overwrite', header=False, sep='\t')
```

```python
# Stop the SparkSession
spark.stop()
```
